```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose


df = pd.read_excel("season-data.xlsx",engine='openpyxl', sheet_name="SAMPLE", index_col="data")

df = df["2012":"2022"]

result = seasonal_decompose(
    
    df['actual'], 
    model='mul', 
    # period=12,
    extrapolate_trend=False,
    two_sided=False
    
    )

df['trend'] = result.trend
df['seasonal'] = result.seasonal
df['resid'] = result.resid

df[['factor', 'seasonal']].plot(figsize=(12,6));
```
